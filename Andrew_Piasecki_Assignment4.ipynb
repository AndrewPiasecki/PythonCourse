For the following questions, the file data_extended.csv is provided, which contains a list of stocks (similar to S&P 500) and certain data about each. Use pandas read_csv() to load the file into a DataFrame, and for each part below, develop statements to further refine the data.

```python
import pandas as pd

# Assuming the file is in the same directory as your script
df = pd.read_csv('/content/sample_data/data_extended.csv')
df
```

a) SEC filings: Use the drop() method to take out the SEC filings column from the DataFrame.

```python
# a) Drop the SEC filings column

df = df.drop('SEC Filings', axis = 'columns')
df
```

b) Headquarters location: Use the split() method to create two new fields, one for city and one for state.

```python
# b) Split Headquarters Location into city and state

df[['City', 'State']] = df['Headquarters Location'].str.split(',', expand = True)
df
```


c) Date first added missing values: Create a DataFrame that has the symbols with those stocks where the Date first added field is missing a value.

```python
# c) Date first added missing values
missing_date_added =df[df['Date First Added'].isnull()]
missing_date_added
```

d) Date first added (multivalued): Find rows where “Date first added” is different from a single date, then use statements to create a new column that determines and populates the date in first position if more than one date is indicated for this field.

```python
# d) Date first added (multivalued)
# Assuming the separation by semicolon for multivalued fields

def firstDate(date):
  if isinstance(date, str):
    return date.split(';')[0]
  else:
    return date

df['First Date Added'] = df['Date First Added'].apply(firstDate)
df
```
